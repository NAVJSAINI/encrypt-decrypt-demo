```python
from cryptography.fernet import Fernet

# Generate a key
key = Fernet.generate_key()
cipher = Fernet(key)

# Input message
message = b"Hello, this is my secret message!"

# Encrypt the message
encrypted_message = cipher.encrypt(message)

# Decrypt the message
decrypted_message = cipher.decrypt(encrypted_message)

print("----- SYMMETRIC ENCRYPTION -----")
print("Key:", key)
print("Original Message:", message)
print("Encrypted Message:", encrypted_message)
print("Decrypted Message:", decrypted_message)
```

```python
from cryptography.hazmat.primitives.asymmetric import rsa, padding
from cryptography.hazmat.primitives import hashes

# Generate private key
private_key = rsa.generate_private_key(
    public_exponent=65537,
    key_size=2048
)

# Generate public key
public_key = private_key.public_key()

# Message
message = b"Asymmetric encryption example!"

# Encrypt with public key
encrypted_message = public_key.encrypt(
    message,
    padding.OAEP(
        mgf=padding.MGF1(algorithm=hashes.SHA256()),
        algorithm=hashes.SHA256(),
        label=None
    )
)

# Decrypt with private key
decrypted_message = private_key.decrypt(
    encrypted_message,
    padding.OAEP(
        mgf=padding.MGF1(algorithm=hashes.SHA256()),
        algorithm=hashes.SHA256(),
        label=None
    )
)

print("\n----- ASYMMETRIC ENCRYPTION -----")
print("Original Message:", message)
print("Encrypted Message:", encrypted_message)
print("Decrypted Message:", decrypted_message)
```

```python
# Simple user database
users = {
    "admin": {"password": "1234", "role": "admin"},
    "user": {"password": "abcd", "role": "user"}
}

# Ask for login
username = input("Enter username: ")
password = input("Enter password: ")

# Check login
if username in users and users[username]["password"] == password:
    print("Login successful!")

    role = users[username]["role"]

    if role == "admin":
        print("Access granted: Admin features")
    else:
        print("Access granted: User features")
else:
    print("Invalid login")
```
